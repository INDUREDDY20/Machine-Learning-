5. ML Model 

```python
x = tc[['Age_08_04','KM','HP','Weight']]
y = tc['Price']
```

```python
x
```

```python
y
```

5.1 Model Building and Validation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=35)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
```

```python
mlr.fit(x_train,y_train)
```

```python
y_pred_train = mlr.predict(x_train)
y_pred_train
```

```python
y_pred_test = mlr.predict(x_test)
y_pred_test
```

```python
error_train = y_train - y_pred_train
error_train
```

```python
error_test = y_test - y_pred_test
error_test
```

Model Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,root_mean_squared_error,r2_score,mean_squared_error
```

```python
r2_train = r2_score(y_train,y_pred_train)
r2_test = r2_score(y_test,y_pred_test)
print(r2_train,r2_test)
```

```python
MAE=mean_absolute_error(y_test,y_pred_test)
MAE
```

```python
MAPE = mean_absolute_percentage_error(y_test,y_pred_test)
MAPE
```

```python
MSE = mean_squared_error(y_test,y_pred_test)
MSE
```

```python
RMS = root_mean_squared_error(y_test,y_pred_test)
RMS
```

Homoscedsicity check:

```python
plt.scatter(x = x_test['HP'],y = error_test)
plt.title('Homoscedsicity check - Scaled_HP vs Error')
plt.xlabel('Scaled_HP')
plt.ylabel('Error')
plt.show()
```

```python
plt.scatter(x = x_test['KM'],y = error_test)
plt.title('Homoscedsicity check - Scaled_KM vs Error')
plt.xlabel('Scaled_KM')
plt.ylabel('Error')
plt.show()
```
