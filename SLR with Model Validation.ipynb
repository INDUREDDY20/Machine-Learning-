```python
npd = pd.read_csv(r"C:\Users\INDU PRIYA\OneDrive\Attachments\Desktop\ML Models\NewspaperData.csv")
npd
```

```python
del npd ["Newspaper"]
npd
```

```python
x = npd[['daily']]
y = npd['sunday']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=16,shuffle=True)
```

```python
x_train,x_test,y_train,y_test
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
```

```python
slr.fit(x_train,y_train)
```

```python
y_pred_train = slr.predict(x_train)
y_pred_train
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_pred_train)
```

```python
y_pred_test = slr.predict(x_test)
y_pred_test
```

```python
r2_score(y_test,y_pred_test)
```
