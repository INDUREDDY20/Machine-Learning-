```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler,MaxAbsScaler
```

```python
minmax_df = pd.DataFrame({
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.randint(15000, 250000, 100),
    'Experience_Years': np.random.randint(0, 35, 100),
    'Score': np.random.randint(30, 100, 100)
})
```

MinMax Scaling

```python
minmax_df.head()
```

```python
mms = MinMaxScaler().fit_transform(minmax_df)
```

```python
mms_df = pd.DataFrame(mms, columns=minmax_df.columns)
```

```python
mms_df.head()
```

Standard Scalar

```python
standard_df = pd.DataFrame({
    'Temperature': np.random.normal(25, 5, 100),
    'Humidity': np.random.normal(60, 10, 100),
    'Pressure': np.random.normal(1015, 8, 100),
    'Wind_Speed': np.random.normal(10, 3, 100)
})
```

```python
standard_df.head()
```

```python
ss =  StandardScaler().fit_transform(standard_df)
```

```python
ss_df = pd.DataFrame(ss,columns=standard_df.columns)
```

```python
ss_df.head()
```

Robust Scaling 

```python
robust_data = np.random.normal(500, 30, 95)
outliers = np.array([2000, 2500, 3000, 3500, 4000])

robust_df = pd.DataFrame({
    'Transaction_Amount': np.concatenate([robust_data, outliers])
})
```

```python
robust_df
```

```python
rs = RobustScaler().fit_transform(robust_df)
```

```python
rs_df = pd.DataFrame(rs,columns=robust_df.columns)
```

```python
rs_df.head()
```

Max Absolute Scalar

```python
maxabs_df = pd.DataFrame({
    'Signal_1': np.random.randint(-300, 300, 100),
    'Signal_2': np.random.randint(-500, 500, 100),
    'Signal_3': np.random.randint(-800, 800, 100)
})
```

```python
maxabs_df.head()
```

```python
mas = MaxAbsScaler().fit_transform(maxabs_df)
```

```python
mas_df = pd.DataFrame(mas,columns=maxabs_df.columns)
```
